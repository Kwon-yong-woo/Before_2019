```python
import requests
from urllib.request import *
from bs4 import BeautifulSoup
import csv
import time
import pandas as pd
import random
import re
```

```python
def get_contents(url):
    f=urllib.request.urlopen(url)
    contentsByte=f.read()
    time.sleep(2)
    try:
        contentsStr=contentsByte.decode('euc-kr')
    except:
        try:
            contentsStr=contentsByte.decode('cp949')
        except:
            contentsStr=contentsByte.decode('UTF-8')
        #try:
        #    contentsStr=contentsByte.decode('utf-8')
    f.close()
    return contentsStr
```

```python
#-*- coding: utf-8 -*-
#파일 열기
fin = open('ihanex_article_links.txt', 'r')

#데이터베이스 리스트 설정 - all data
fline = fin.readline().strip()
all_data=[]
count=0
while fline !="":
    count = count+1
    #게시글 정보 리스트 설정 - article_data
    article_data=[]
    #글 번호 파싱
    numbsrch = re.search('no=(.+)&', fline)
    numb = numbsrch.group(1)
    article_data.append(numb)
    time.sleep(1)
    url = urlopen(fline)
    time.sleep(1)
    bsobj = BeautifulSoup(url, 'html.parser')
    time.sleep(1.5)
    sbsobj = str(bsobj)
    #글 날짜 파싱
    #cont = urllib.request.urlopen(fline).read()
    #contents = cont.decode('euc-kr')
    m = re.search(": ([\d-]+)", sbsobj)
    naljja = m.group(1)
    article_data.append(naljja)
    #글 제목 파싱    
    title = bsobj.find_all("font", {"class" : "view_title2"})
    for t in title:
        article_data.append(t.get_text().lstrip().rstrip())
    #글 내용 파싱
    contents = bsobj.find_all('td', {"class" : 'board-contents'})
    for c in contents:
        texxt = c.get_text().lstrip()
        texxt = texxt.replace(u'\xa0', ' ')
        texxt = texxt.replace(u'\n', ' ')
        article_data.append(texxt.rstrip())
    #fout.writerow(article_data)
    for text in bsobj.select('tr > td > table > tbody > tr'):
        print(text.get_text().lstrip())
    all_data.append(article_data)
    print(numb)
    if len(all_data) % 50 == 0:
        print('현재 총 {0}개 처리'.format(count))
        time.sleep(10)
    #if count == 400:
    #    break
    fline = fin.readline().strip()
fin.close()
```
